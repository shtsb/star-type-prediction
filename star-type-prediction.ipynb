```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

s = pd.read_csv("6 class csv.csv")
s.dropna(inplace=True)
s.drop_duplicates(inplace=True)

s = s[["Temperature (K)","Luminosity(L/Lo)","Radius(R/Ro)","Absolute magnitude(Mv)","Star type"]]
X = np.array(s.drop(["Star type"],axis=1))
Y = np.array(s["Star type"])

X_train , X_test , Y_train , Y_test = train_test_split(X,Y,test_size=0.2)

model = LinearRegression()
model.fit(X_train,Y_train)

predicted= model.predict(X_test)

for i in range(len(predicted)):
    print (f"real:{Y_test[i]} , predicted : {round(predicted[i])}")


acc = model.score(X_test,Y_test)
print("\n\naccuracy :", acc)
print("coef(weight):",model.coef_)
print("intercept(bias):",model.intercept_)

plt.plot(Y_test,label="test")
plt.plot(predicted,label="predicted")
plt.suptitle("star type prediction")
plt.legend()
plt.show()

```
